```python
t = np.linspace( -3, 3, 51)
y = np.cos(t)
                
```

```python
plt.plot(t, y)
plt.show()
```

```python

title_dict = { 'fontsize':20, 'fontweight':'bold', 'color':'#808080', 'family':'calibri'} 
plt.plot(t, y)
plt.title('График f(x)', fontdict = title_dict, loc = 'left')

plt.xlabel('Переменная t')
plt.ylabel('Значение Функции F(t)')

plt.show()
```

```python
x = np.linspace( -3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = 3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols =4 )
ax1, ax2, ax3, ax4 = ax
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
```

```python
fig, ax = plt.subplots(nrows=2, ncols = 2 )
ax1, ax2, ax3, ax4 =ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
fig.set_size_inches(8,6)
plt.subplots_adjust( wspace = 0.3, hspace = 0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols = 2 )
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title ('График  x**2')
ax1.set_xlim([-1, 1])

ax2.plot(x, y2)
ax2.set_title ('График 2*x + 0.5')

ax3.plot(x, y3)
ax3.set_title ('График 3*x - 1.5')

ax4.plot(x, y4)
ax4.set_title ('График sin(x)')

fig.set_size_inches(8,6)
plt.subplots_adjust( wspace = 0.3, hspace = 0.3)
```

```python
plt.figure(figsize=(12, 8))

c = df['Class'].value_counts()
print(type(c))
print(c[0])
ax = df['Class'].value_counts().plot(kind='bar', logy = True)
ax.set_title('Колчисвто Class')
ax.set_xlabel('Значение измерения Class')
ax.set_ylabel('Количество измерений')

rects = ax.patches
for rect, label in zip(rects, c):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label,
            ha='center', va='bottom')
```
